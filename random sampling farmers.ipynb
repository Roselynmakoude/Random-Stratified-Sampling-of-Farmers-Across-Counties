ASSIGNMENT: CREATE A RANDOM SAMPLE OF 120 FARMERS FROM THE 3 COUNTIES. 

Each county should have a sample of 40 farmers based on the criteria below.
-Sex of farmer (male/female)
-Age – breakdown by 17-29 (sample of both male and female), and 30-65 (both male and female)
-Clustering per village (villages with low, medium and high concentration of farmers)
-Cohorts (referred to in Excel sheet as “instance”) – sampling across cohorts (when farmers were recruited)

```python
#Import necessary libraries
import pandas as pd
```

```python
data=pd.read_csv("C:\\Users\\Administrator\\OneDrive\\Documents\\List of Farmer for Sample Size Selection- Quality Evaluation_Confidential.csv")
data.head()
```

```python
#Part 1: Clustering per village
village_counts=data["Village"].value_counts()
#define concentration
high_conc=village_counts.quantile(0.66)
low_conc=village_counts.quantile(0.33)

def village_clusters(x):
    if village_counts[x]>=high_conc:
        return "high"
    elif village_counts[x]>=low_conc:
        return "low"
    else:
        return "medium"

data["village_clusters"]=data["Village"].map(village_clusters)
```

```python
#Part 2: Creating age groups broken down by gender
bins=[17,29,65]
labels=["17-29", "30-65"]
data["age_group"]=pd.cut(data["Age"], bins=bins, labels=labels, right=True, include_lowest=True)
data=data.dropna()
data
```

```python
#check for null values
data.isna().sum()
```

```python
#change the data type
data["age_gender_group"]=data["Gender"].astype(str) + "_" + data["age_group"].astype(str)
```

```python
#Splitting the County column
data[["country", "rest"]]=data["Project Unit Name"].str.split(":", n=1, expand=True)
data
```

```python
data["country"] = data["country"].str.strip()
data["rest"] = data["rest"].str.strip()
```

```python
data[["County", "zone_no"]]=data["rest"].str.rsplit(" ", n=1, expand=True)
data.head()
```

```python
#Drop unnecessary columns
data=data.drop(columns=["Project Unit Name", "country", "rest", "zone_no"])
data.head()
               
```

```python
data["County"] = data["County"].replace({"Homa Bay":"HomaBay"})
data
```

CREATING THE RANDOM SAMPLE

```python
#Building a strata key
data["strata"]=(
    data["County"].astype(str) + "_"+
    data["age_gender_group"].astype(str) + "_"+
    data["Instance"].astype(str)+ "_" +
    data["village_clusters"].astype(str)
)
```

```python
print("strata per County:")
print(data.groupby("County")["strata"].nunique())

print("\nCounts per County:")
print(data["County"].value_counts())
```

```python
print(data["County"].unique())      
print(data["strata"].nunique())         
print(data["County"].value_counts())
```

```python
#Part 5: Proportional Stratified Sampling
samples = []

for County, subset in data.groupby("County"):
    target_n = 40
    
    counts = subset["strata"].value_counts()
    proportions = counts / counts.sum()
    n_per_strata = (proportions * target_n).round().astype(int)

    diff = target_n - n_per_strata.sum()
    if diff != 0:
        n_per_strata.iloc[0] += diff

    strat_samples = []
    for strata, n in n_per_strata.items():
        group = subset[subset["strata"] == strata]
        if len(group) > 0 and n > 0:
            strat_samples.append(group.sample(n=min(n, len(group)), random_state=42))

    if strat_samples:
        samples.append(pd.concat(strat_samples))


if samples:
    random_sample = pd.concat(samples, ignore_index=True)
    print(random_sample.shape)
    print(random_sample["County"].value_counts())
```
